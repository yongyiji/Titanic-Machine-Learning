```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_train = pd.read_csv("/Users/jiyi/Desktop/534/project2/train.csv")
data_train
```

```python
data_train=data_train.drop(['Name', 'Ticket','PassengerId','Cabin'], axis=1)
data_train
```

```python
data_train=data_train.dropna(axis=0)
data_train
```

```python
data_train = pd.get_dummies(data_train, columns = ['Sex'], drop_first=True)
data_train= pd.get_dummies(data_train, columns = ['Embarked'], drop_first=True)
data_train.head(10)
```

```python
X_train = data_train[['Pclass','Age','SibSp','Parch','Fare','Sex_male','Embarked_Q','Embarked_S']]
y_train = data_train[['Survived']]
print(X_train) 
print(y_train)
```

```python
from sklearn.model_selection import train_test_split
X_train_train, X_train_test, y_train_train, y_train_test = train_test_split(X_train,y_train, test_size = 0.2,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler 
scaler = StandardScaler() 
X_train_scale=scaler.fit_transform(X_train_train[['Pclass','Fare','Age','SibSp','Parch']]) 
X_test_scale=scaler.transform(X_train_test[['Pclass','Fare','Age','SibSp','Parch']])
```

```python
X_train_train['Pclass']=X_train_scale[:,0]
X_train_train['Fare']=X_train_scale[:,1]
X_train_train['Age']=X_train_scale[:,2]
X_train_train['SibSp']=X_train_scale[:,3]
X_train_train['Parch']=X_train_scale[:,4]
```

```python
X_train_train
```

```python
X_train_test['Pclass']=X_test_scale[:,0]
X_train_test['Fare']=X_test_scale[:,1]
X_train_test['Age']=X_test_scale[:,2]
X_train_test['SibSp']=X_test_scale[:,3]
X_train_test['Parch']=X_test_scale[:,4]
```

```python
X_train_test
```

kernel SVM

```python
from sklearn.svm import SVC
SVM_rbf = SVC(kernel="rbf",random_state=0) 
SVM_rbf.fit(X_train_train, y_train_train)
```

```python
pred_SVM_rbf=SVM_rbf.predict(X_train_test)
```

```python
y_train_test_SVM_rbf=y_train_test
y_train_test_SVM_rbf['Predict Data'] = pred_SVM_rbf.tolist() 
y_train_test_SVM_rbf.head(10)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score 
cm_SVM_rbf=confusion_matrix(y_train_test_SVM_rbf['Survived'].values,pred_SVM_rbf)
cm_SVM_rbf
```

```python
as_SVM_rbf=accuracy_score(y_train_test_SVM_rbf['Survived'].values,pred_SVM_rbf)
as_SVM_rbf
```

```python
SVM_poly = SVC(kernel="poly",random_state=0) 
SVM_poly.fit(X_train_train, y_train_train)
```

```python
pred_SVM_poly=SVM_poly.predict(X_train_test)
```

```python
y_train_test_SVM_poly=y_train_test
y_train_test_SVM_poly['Predict Data'] = pred_SVM_poly.tolist() 
y_train_test_SVM_poly.head(10)
```

```python
cm_SVM_poly=confusion_matrix(y_train_test_SVM_poly['Survived'].values,pred_SVM_poly)
cm_SVM_poly
```

```python
as_SVM_poly=accuracy_score(y_train_test_SVM_poly['Survived'].values,pred_SVM_poly)
as_SVM_poly
```

LR

```python
from sklearn.linear_model import LogisticRegression 
Logistic = LogisticRegression(random_state=0) 
Logistic.fit(X_train_train, y_train_train)
```

```python
pred_Logistic=Logistic.predict(X_train_test)
```

```python
y_train_test_Logistic=y_train_test
y_train_test_Logistic['Predict Data'] = pred_Logistic.tolist() 
y_train_test_Logistic.head(10)
```

```python
cm_Logistic=confusion_matrix(y_train_test_Logistic['Survived'].values,pred_Logistic)
cm_Logistic
```

```python
as_Logistic=accuracy_score(y_train_test_Logistic['Survived'].values,pred_Logistic)
as_Logistic
```

```python
from sklearn.neighbors import KNeighborsClassifier
KNN = KNeighborsClassifier(n_neighbors=5,metric="minkowski") 
KNN.fit(X_train_train, y_train_train)
```

```python
pred_KNN=KNN.predict(X_train_test)
```

```python
y_train_test_KNN=y_train_test
y_train_test_KNN['Predict Data'] = pred_KNN.tolist() 
y_train_test_KNN.head(10)
```

```python
as_KNN=accuracy_score(y_train_test_KNN['Survived'].values,pred_KNN)
as_KNN
```

```python
from sklearn.tree import DecisionTreeClassifier
DT = DecisionTreeClassifier(criterion = 'entropy', random_state = 0) 
DT.fit(X_train_train, y_train_train)
```
